```python
# importando as bibliotecas necessárias
import pandas as pd
import pydeck as pdk
```

```python
# importando os dados
df = pd.read_csv("localidades_1gav9_2021.csv")
origem_destino = pd.read_csv("origem_destino_2021.csv")
```

```python
# para evitar um excesso de informação no mapa, foram selecionadas apenas as localidades que aparecem como destino por mais de 15 vezes
selecao_localidades = origem_destino.destino.value_counts()
selecao = selecao_localidades[origem_destino.destino.value_counts() >= 15]
selecao = pd.DataFrame(selecao)


x = origem_destino.loc[origem_destino.destino.isin(selecao.index)]
x = x.loc[x.origem.isin(selecao.index)]
x = x.loc[x.destino != "SBMN"]
x = x.loc[~x.origem.isin(["SWUQ","SWBV","SBYA","SWQE","SWSQ","SWMK","SWJP","SWEE","SWII","SWEI","SWLB"])]

GREEN_RGB = [0, 255, 0, 40]
RED_RGB = [240, 100, 0, 40]

#mapabrasil

# criando a camada
layer = pdk.Layer( 
    "HeatmapLayer" ,  # `type` positional argument is here
    df,
    get_position=['long', 'lat'],
    elevation_range=[0,10],
    coverage=1,
    auto_highlight=True)

arc_layer = pdk.Layer(
    "ArcLayer",
    data=x,
    get_width="S000 * 2",
    get_source_position=["long_origem", "lat_origem"],
    get_target_position=["long_dest", "lat_dest"],
    get_tilt=15,
    get_source_color=RED_RGB,
    get_target_color=GREEN_RGB,
    pickable=True,
    auto_highlight=True, bearing=0, pitch=10)

# definindo a visão
view_state = pdk.ViewState(
    longitude=-50,
    latitude=-5.3545124,
    zoom=3.5,
    min_zoom=3,
    max_zoom=15, bearing=0, pitch=10)

# combinando os dois em um deck, e visualizando
r = pdk.Deck(layers=[layer,arc_layer], initial_view_state=view_state, map_style='light')

# salvando como html
r.to_html('heatmap1gav9.html')
```
